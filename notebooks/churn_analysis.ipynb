# **Analyzing Customer Churn**

## **Introduction**

This notebook contains a detailed analysis of the customer churn rate for the telecommunications company Databel. Power BI tools were used to create visualizations and calculate key metrics, which allowed us to identify the drivers of churn and propose strategies for its reduction.

The analysis focuses on the following areas:
- Main causes of customer churn.
- Churn distribution by age, state, and payment method.
- Recommendations to improve customer retention.

Below are the key findings obtained from the data analysis.

---

## **1. Overall Churn Rate**

The company’s overall churn rate is **26.86%**, meaning nearly one in four customers leaves the service within a given period. This is a concerning indicator, and the following sections explore the reasons behind this churn.

### **Competition as the Dominant Factor**

- **44.82%** of customers who left did so due to better offers or devices from competitors.  
- This highlights the importance of improving Databel’s product and service competitiveness, as well as its value proposition to customers.

![Churn Rate](../visualizations/global_churn_rate.png)

---

## **2. Churn by Age**

### **Older Age Group (80+) with High Churn**

- Customers over 80 years old have a churn rate of **52%**, much higher than the overall average of 26.86%.  
- As age increases, so does the churn rate, peaking among senior customers. This may indicate a disconnect between the services offered and the specific needs of this group.

![Churn by Age](../visualizations/churn_by_age.png)

---

## **3. Churn by State**

### **California: A High Churn Case with Low Interaction**

- **California** has the highest churn rate at **63.24%**.  
- Despite this high churn, the number of customer service calls is surprisingly low. This suggests that customers are deciding to leave the service without attempting to resolve their issues through support, which presents an opportunity to improve retention.

![Churn by State](../visualizations/churn_by_state.png)

---

## **4. Data Plan Usage**

### **Customers with Unlimited Plans and Low Usage**

- Customers with an unlimited data plan but consuming less than 5 GB of data have a churn rate of **34.71%**, significantly higher than those without an unlimited plan (**12.31%**).  
- This difference shows the importance of tailoring data plans to better fit users' consumption habits.

![Data Plans](../visualizations/data_plans.png)

---

## **Conclusions and Recommendations**

1. **Improve Offer Competitiveness**: The main reason for churn is competition. Databel should enhance its offers and personalize plans to retain customers.

2. **Focus on Senior Customers**: The high churn rate among older customers suggests that services should be adjusted for this segment, offering additional support and personalized products.

3. **Research in California**: California has an alarming churn rate. It is recommended to conduct customer surveys and improve customer service to address the root issues.

4. **Review Unlimited Plans**: Databel should consider adjusting its unlimited plans, especially for customers with low data usage, by offering more economical and flexible options.
