# **Análisis de la Pérdida de Clientes**

## **Introducción**

Este cuaderno contiene el análisis detallado de la tasa de rotación de clientes de la empresa de telecomunicaciones Databel. Se utilizaron herramientas de Power BI para crear visualizaciones y calcular las métricas clave, lo que permitió identificar los factores que impulsan la rotación y proponer estrategias para su reducción.

El análisis se centra en las siguientes áreas:
- Causas principales de la rotación de clientes.
- Distribución de la rotación por edad, estado y método de pago.
- Recomendaciones para mejorar la retención de clientes.

A continuación, se detallan los resultados clave obtenidos a partir del análisis de datos.

---

## **1. Tasa de Rotación Global**

La tasa de rotación general de la empresa es del **26.86%**, lo que significa que casi uno de cada cuatro clientes abandona el servicio en un periodo determinado. Este es un indicador preocupante, y en las siguientes secciones exploramos las razones detrás de esta rotación.

### **Competencia como Factor Dominante**

- El **44.82%** de los clientes que dejaron el servicio lo hicieron debido a mejores ofertas o dispositivos ofrecidos por competidores.  
- Este factor resalta la importancia de mejorar la competitividad de los productos y servicios de Databel, así como la oferta de valor a los clientes.

![Churn Rate](../visualizations/global_churn_rate.png) 

---

## **2. Rotación por Edad**

### **Segmento de Edad Mayor (80+) con Alta Rotación**

- Los clientes mayores de 80 años presentan una tasa de rotación del **52%**, muy superior al promedio general del 26.86%.  
- A medida que aumenta la edad, también aumenta la tasa de rotación, con un pico en los clientes más seniors. Esto puede indicar una desconexión entre los servicios ofrecidos y las necesidades específicas de este grupo.

![Rotación por Edad](../visualizations/churn_by_age.png)

---

## **3. Rotación por Estado**

### **California: Un Caso de Alta Rotación con Baja Interacción**

- **California** es el estado con el índice de rotación más alto, alcanzando un **63.24%**.  
- A pesar de esta alta rotación, el número de llamadas al servicio al cliente es sorprendentemente bajo. Esto sugiere que los clientes están tomando la decisión de abandonar el servicio sin intentar resolver sus problemas con el soporte, lo cual es una oportunidad para mejorar la retención.

![Rotación por Estado](../visualizations/churn_by_state.png)

---

## **4. Uso de Planes de Datos**

### **Clientes con Planes Ilimitados y Baja Consumo**

- Los clientes con un plan de datos ilimitado pero que consumen menos de 5 GB de datos presentan una tasa de rotación del **34.71%**, considerablemente superior a los que no tienen plan ilimitado (**12.31%**).  
- Esta diferencia muestra la importancia de personalizar los planes de datos para ajustarse mejor a los hábitos de consumo de los usuarios.

![Planes de Datos](../visualizations/data_plans.png)

---

## **Conclusiones y Recomendaciones**

1. **Mejorar la Competitividad de las Ofertas**: La principal razón de la rotación es la competencia. Databel debería mejorar sus ofertas y personalizar sus planes para retener a los clientes.
  
2. **Atención a Clientes Senior**: El alto churn rate en clientes mayores sugiere que se deben ajustar los servicios para este segmento, ofreciendo soporte adicional y productos personalizados.

3. **Investigación en California**: California tiene una tasa de rotación alarmante. Se recomienda realizar encuestas a los clientes y mejorar el servicio de atención para abordar los problemas de raíz.

4. **Revisión de Planes Ilimitados**: Databel debe considerar ajustar sus planes ilimitados, especialmente para los clientes que consumen menos datos, ofreciendo opciones más económicas y flexibles.
